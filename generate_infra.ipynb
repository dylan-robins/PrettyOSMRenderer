```python
places = "Grenoble, France"
tags = {'amenity':True, 'landuse':['retail','commercial'], "highway":"bus_stop"}
gdf = ox.geometries_from_place(places, tags)
gdf.shape
fig, ax = ox.plot_footprints(gdf, figsize=(3, 3))
```
